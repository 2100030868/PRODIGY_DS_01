```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("population.csv")
data.head(6)
```

```python
data.tail(7)
```

```python
con = ['Angola','Canada','Brazil','Denmark','Georgia','Zambia']
fd = data[data['Country Name'].isin(con)]
plt.bar (fd['Country Name'], fd['2022'],color='green')
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


con = ['Angola', 'Canada', 'Brazil', 'Denmark', 'Georgia', 'Zambia']

fd = data[data['Country Name'].isin(con)]

plt.hist(fd['2022'], bins=10, edgecolor='black',color='red')
plt.xlabel('Population in 2022')
plt.ylabel('Frequency')
plt.title('Population Distribution of Selected Countries in 2022')
plt.show()
```
